### Bir fonksiyon oluÅŸturun

`city` ve `key` olmak Ã¼zere iki argÃ¼man alan ve ÅŸehir iÃ§in karÅŸÄ±lÄ±k gelen deÄŸeri dÃ¶ndÃ¼ren `get_key_value_from_city` fonksiyonunu oluÅŸturun.

__Ã–rnek__: `get_key_value_from_city(city=paris_city, key="total_population")` 2102650 dÃ¶ndÃ¼recektir.

<u>Not:</u> EÄŸer bir anahtar (key) yoksa, fonksiyon `"_<key>_ not defined"` dÃ¶ndÃ¼rmelidir.
    




````python
def get_key_value_from_city(city, key):
    if key in city:
        return city[key]
    return f"_{key}_ not defined"
````

### DeÄŸerleri okuma


Ã–nceki `get_key_value_from_city` fonksiyonunuzu kullanarak iki argÃ¼man alan yeni bir fonksiyon oluÅŸturun: `city_name` ve `key`.

EÄŸer ÅŸehir listede yoksa, fonksiyon `"*city* doesn't exist or is not covered"` dÃ¶ndÃ¼rmelidir.

Fonksiyonunuzu Ã§eÅŸitli Ã¶rneklerle test edin.

````python
def _find_city(cities, city_name):
    for c in cities:
        if c.get("name") == city_name:
            return c
    return None
````

````python
def get_key_value_from_cities(cities, city_name, key):
    city = _find_city(cities, city_name)
    if city is None:
        return f"{city_name} doesn't exist or is not covered"
    return get_key_value_from_city(city, key)
````

````python
get_key_value_from_cities(cities, "Paris", "total_population")
get_key_value_from_cities(cities, "Atlantis", "total_population")
get_key_value_from_cities(cities, "Paris", "foo")
````

### DeÄŸer gÃ¼ncelleme


Bazen ÅŸehir bilgileri yanlÄ±ÅŸ olabilir veya gÃ¼ncellemeye ihtiyaÃ§ duyabilir.

`modify_city_key_value` fonksiyonunu oluÅŸturun; bu fonksiyon mevcut bir anahtarÄ±n deÄŸerini yeni bir deÄŸerle deÄŸiÅŸtirir. \
Fonksiyon Ã¼Ã§ argÃ¼man alÄ±r: `city_name`, `key` ve anahtara atanacak yeni `value`.

Fonksiyon aÅŸaÄŸÄ±daki durumlardan birini belirten bir metin mesajÄ± Ã¼retecektir:

- EÄŸer ÅŸehir yoksa veya kapsanmÄ±yorsa: `"*city* doesn't exist or is not covered."`
- EÄŸer verilen anahtar yoksa: `"*key* doesn't exist."`
- EÄŸer anahtar baÅŸarÄ±yla gÃ¼ncellendiyse: `"*key* has been updated for *city*."`

````python
def modify_city_key_value(cities, city_name, key, value):
    city = _find_city(cities, city_name)
    if city is None:
        return f"{city_name} doesn't exist or is not covered."
    if key not in city:
        return f"{key} doesn't exist."
    city[key] = value
    return f"{key} has been updated for {city_name}."
````

````python
modify_city_key_value(cities, "Paris", "total_population", 999)
get_key_value_from_cities(cities, "Paris", "total_population")
modify_city_key_value(cities, "Paris", "foo", 1)
modify_city_key_value(cities, "Atlantis", "total_population", 1)
````

### Arama motorunu tasarlayÄ±n


`get_cities_nearby` adlÄ± bir fonksiyon oluÅŸturun; bu fonksiyon iki argÃ¼man alÄ±r:
- `customer_location`: [lat, long] formatÄ±nda koordinatlardan oluÅŸan bir liste


- `radius`: kilometre cinsinden eÅŸik deÄŸeri

Ve verilen `customer_location` yakÄ±nÄ±nda bulunan `ÅŸehirlerin bir listesini` dÃ¶ndÃ¼rÃ¼r.

EÄŸer hiÃ§bir ÅŸehir bulunamazsa, fonksiyon boÅŸ bir liste dÃ¶ndÃ¼rmelidir.

Test etmek iÃ§in Ã‡in'den Guangyan ÅŸehrini ([32.4355, 105.8436]) ve 500 km'lik bir yarÄ±Ã§apÄ± kullanabilirsiniz.

Xi'an ve Chengdu'nun Ã§ok uzak olmadÄ±ÄŸÄ±nÄ± gÃ¶rmelisiniz...

````python
def get_cities_nearby(cities, customer_location, radius):
    nearby = []
    for city in cities:
        city_loc = city.get("location")
        if not city_loc or len(city_loc) != 2:
            continue
        d_km = distance.distance(customer_location, city_loc).km
        if d_km <= radius:
            nearby.append(city["name"])
    return nearby
````

````python
get_cities_nearby(cities, [32.4355, 105.8436], 500)
````

# Travel Agency Database Project

This project is a small but complete Python application that simulates how a travel agency can store, read, update, and query city information using core Python data structures.

I built this project to practice working with dictionaries and lists in a realistic scenario, to design reusable functions, and to integrate a real-world library for geographic distance calculations.

The entire implementation is done inside a Jupyter Notebook for clarity and step-by-step validation.

---

## Project Overview

The project represents a travel agency database where each city is stored as a Python dictionary and all cities are collected inside a list called `cities`. Each city contains basic information such as name, population, continent, geographic coordinates, and administrative area data.

The goal was not just to store data, but to actively interact with it through well-defined functions that handle both normal and edge cases.

---

## What I Implemented

I started by working with a single city dictionary to understand nested data access. I retrieved values from lists and dictionaries, checked data types, and safely accessed deeply nested keys.

I then created a function that reads a value from a city dictionary given a key. If the key does not exist, the function returns a clear and controlled message instead of failing.

After that, I expanded the logic to work with multiple cities stored in a list. I implemented a helper function to locate a city by name and used it to build higher-level functions that can retrieve or update city data by city name and key. These functions explicitly handle cases where the city does not exist or the key is invalid.

In the final stage of the project, I integrated the `geopy` library to perform real geographic distance calculations. Using latitude and longitude coordinates, I implemented a search function that returns all cities within a given radius of a customerâ€™s location. This turned the project from a simple data exercise into a realistic travel search feature.
